Since we've established that there are missing values, let's deal with them

The first step is to check for place holders and remove them

```python
# Common placeholders to check for
placeholders = ["null","Null" ,  "NULL", "NA", "N/A", "NaN", "nan", "None", "none", "?", "-", ""]

# Check which cells contain these placeholders
for placeholder in placeholders:
    mask = df.map(lambda x: str(x).strip() == placeholder)
    print(f"'{placeholder}' found in {mask.sum().sum()} cells")
```

since there are some place holders we need to remove them

```python
df.replace(placeholders, np.nan, inplace=True)
df_cleaned = df.dropna()
```

Some columns have too many missing values like the air carrier column and the Schedule column. 
we need to drop them

```python
df.drop(['Air carrier', 'Schedule'], axis=1, inplace=True, errors='ignore')
```

```python
df.columns
```

now lets deal with the numeric columns first and fill them with 0


```python
#from the statistical report, we already see all the numerical columns
numerical_columns = [
    'Total Fatal Injuries', 'Total Serious Injuries',
    'Total Minor Injuries', 'Total Uninjured', 'Number of Engines'
]
for columns in numerical_columns :
    df[columns] = df[columns].fillna(0).astype(int)
```

We also need to ensure that date, latitude and longitude have the correct data types

```python
#Correct Data Types for Date and Lat/Long
df['Event Date'] = pd.to_datetime(df['Event Date'], errors='coerce')
df['Publication Date'] = pd.to_datetime(df['Publication Date'], dayfirst= False)


df['Latitude'] = pd.to_numeric(df['Latitude'] ,  errors='coerce')
df['Longitude'] = pd.to_numeric(df['Longitude'],  errors='coerce')
```

Without forgetting the categorical data, we are going to fill them with 'unknown'

```python
 #Impute Categorical columns with missing Values with 'Unknown'
categorical_columns = [
    'Aircraft Category', 'Airport Code', 'Airport Name',
   'Broad phase of flight', 'Report Status', 'Engine Type',
    'Purpose of flight', 'Weather Condition', 'Aircraft damage',
    'Injury Severity', 'Registration Number', 'Country',
    'Amateur Built', 'Model', 'Make', 'Location', 'Longitude',
    'Latitude', 'Publication Date', 'FAR Description'
 
]
for columns in categorical_columns:
    # Ensure column is of object type before filling with 'Unknown' string, if it's not already
    if df[columns].dtype != 'object':
        df[columns] = df[columns].astype(str)
    df[columns] = df[columns].fillna('Unknown')

categorical_columns
```

```python
# Standardize 'AmateurBuilt' column to 'Yes'/'No'
df['Amateur Built'] = df['Amateur Built'].astype(str).str.title()
df['Amateur Built'] = df['Amateur Built'].replace({'Y': 'Yes', 'N': 'No'})
```

now lets re-check for missing values

```python
df.isna().sum().sort_values(ascending= True)
```

```python
private_use = [
    'Personal', 'Instructional', 'Business', 'Ferry', 'Positioning',
    'Flight Test', 'Airshow', 'Public Aircraft'
]
commercial_use = [
    'Air Taxi', 'Commercial', 'Corporate', 'Aerial Application',
    'Aerial Observation', 'Executive', 'Other Work Use'
]
# Function to categorize purpose of flight
def categorize_purpose(purpose):
    purpose_lower = str(purpose).lower()
    if any(keyword.lower() in purpose_lower for keyword in private_use):
        return 'Private/Personal Use'
    elif any(keyword.lower() in purpose_lower for keyword in commercial_use):
        return 'Commercial/For-Hire Use'
    else:
        return 'Other/Unknown Use'

df['Purpose of flight grouped'] = df['Purpose of flight'].apply(categorize_purpose)
df['Purpose of flight grouped']
```

```python
# First, ensure your DataFrame is properly prepared
print("DataFrame shape:", df.shape)
print("Columns:", df.columns.tolist())

# Create the grouped purpose column 
df['Purpose of flight grouped'] = df['Purpose of flight'].apply(categorize_purpose)

# Check value counts
print("Broad phase of flight value counts:")
print(df['Broad phase of flight'].value_counts())

print("\nPurpose of flight grouped value counts:")
print(df['Purpose of flight grouped'].value_counts())

# Create phase order
phase_order = df['Broad phase of flight'].value_counts().index.tolist()

# Create the visualization
plt.figure(figsize=(14, 8))
sns.countplot(
    data=df,  
    x='Broad phase of flight',
    hue='Purpose of flight grouped',  
    palette='viridis',
    order=phase_order  
)

plt.title('Incident Count by Broad Phase of Flight and Purpose of Flight', fontsize=16)
plt.xlabel('Broad Phase of Flight', fontsize=12)
plt.ylabel('Number of Incidents', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Purpose of Flight', bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```

Incident Count by Broad Phase of Flight and Purpose of Flight Grouped.
This chart identifies at which stages of a flight incidents most frequently occur, again distinguishing between private/personal and other flight purposes.

Analysis:

High-Risk Flight Phases: 'Landing', 'Takeoff', and 'Cruise' consistently represent the highest number of incidents among known flight phases. 'Maneuvering' and 'Approach' are also significant phases for incidents. 


Operational Risk in Private Aviation: Similar to aircraft categories, 'Private/Personal Use' operations consistently show a markedly higher number of incidents across almost all phases of flight, particularly during 'Landing', 'Takeoff', 'Cruise', and 'Maneuvering'. This highlights that fundamental operational phases and specific maneuvers are key risk points, especially for private aviation activities.

Business Recommendations for the New Aviation Division
Based on our analysis of historical NTSB data, here are three concrete recommendations to guide your company's new aviation venture towards lower risk:

1. Prioritize Commercial Operations and Robust Pilot Training for Private Ventures:

Recommendation: Given that 'Private/Personal Use' operations consistently show significantly higher incident counts across almost all aircraft categories and phases of flight, your company should consider initially focusing investments and operations on commercial (for-hire) aviation ventures where regulations are typically more stringent and professional pilot experience is standard.

For Private Operations: If private aircraft operations are pursued, implement exceptionally robust pilot training programs that exceed minimum regulatory requirements, with a strong emphasis on proficiency in critical phases of flight (takeoff, landing, cruise, maneuvering) and emergency procedures. Consider mandating regular recurrent training and flight reviews.

2. Strategic Aircraft Acquisition: Consider Popular & Well-Supported Models, Prioritize Maintenance:

Recommendation: While popular aircraft makes like Cessna, Piper, and Beech show high absolute incident numbers, this correlates with their sheer prevalence. For initial purchases, consider acquiring commonly used models from these established manufacturers. They benefit from wider parts availability, a larger pool of qualified mechanics, and extensive pilot training infrastructure, which can contribute to operational safety and reduce downtime.

Mitigation Strategy: Regardless of make, establish a top-tier maintenance program for all acquired aircraft. Regular, thorough maintenance, going beyond minimum requirements, is a critical factor in mitigating mechanical failures that can lead to incidents.

3. Focus on High-Risk Flight Phases in Operational Protocols:

Recommendation: Develop and enforce rigorous standard operating procedures with specific emphasis on high-risk flight phases: Takeoff, Landing, Cruise, Approach, and Maneuvering.

Implementation: For both commercial and private flights, ensure pilots receive specialized training and adhere to strict checklists during these phases. Consider integrating advanced avionics and automation where appropriate to reduce pilot workload and potential for human error during these critical moments. Regularly review incident reports for these phases to continuously refine and improve operational protocols.
